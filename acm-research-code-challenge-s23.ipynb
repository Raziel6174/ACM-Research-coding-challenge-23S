```python
#library imports
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib as mp # for graphing data
import matplotlib.pyplot as plt

#Lists files under the input directory
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
#Name: Michael Figueroa
#Date: 1-30-2023
#Description: This code is meant to take the information from the star data set and 
#reimagine the Hertzprung Russell Diagram in a 3-D diagram with some additional 
#visualization of information.

star_data = pd.read_csv("/kaggle/input/star-dataset/6 class csv.csv", header = 0, sep = ",")
star_list = star_data.to_dict('list') #Turns dataframe into list
#print(star_list, "\n") #Prints data

#These are the star stats available to graph, Feel free to switch them out in the 
#variables to makes some interesting visualizations!
# -Temperature (K)
# -Luminosity(L/Lo)
# -Radius(R/Ro)
# -Absolute magnitude(Mv)
# -Star type
# -Star color
# -Spectral Class

#Reminder on star type:
#Brown Dwarf -> Star Type = 0
#Red Dwarf -> Star Type = 1
#White Dwarf -> Star Type = 2
#Main Sequence -> Star Type = 3
#Supergiant -> Star Type = 4
#Hypergiant -> Star Type = 5

#Establishes x, y, z, dot size, and color for the graph
x = 'Absolute magnitude(Mv)'
y = 'Star type'
z = 'Temperature (K)'
size = 15
color = 'Luminosity(L/Lo)'

#Build 3D scatter plot
plt.figure(figsize=(10,10))
plt.style.use('_mpl-gallery')
ax = plt.axes(projection='3d')
fig = ax.scatter3D(star_list[x], star_list[y], star_list[z], 
                   s = size, c = star_list[color], cmap = 'RdYlBu', alpha = 1)

#Insert axis labels
ax.set_xlabel(x)
ax.set_ylabel(y)
ax.set_zlabel(z)

#Insert graph colors
cbar = plt.colorbar(fig)
cbar.set_label(color)
```
